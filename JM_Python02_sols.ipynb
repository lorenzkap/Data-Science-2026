# Conditionals<a name="Conditionals"></a>

"Conditionals" is a fancy word for if-statements. If you've ever done any programming, you are surely aware of the if-then-else construction. In Python it's done as follows:



````python
number_of_apples = 5

# The computer starts here, if the following statement is true, it executes the block below
if number_of_apples < 1:  # number_of_apples is NOT <1, therefore, continue
    print('You have no apples')
elif number_of_apples == 1:  # number_of_apples is NOT =1, therefore, continue
    print('You have one apple')
elif number_of_apples < 4:  # number_of_apples is NOT <4, therefore, continue
    print('You have a few apples')
# none of the earlier conditions were true!  Therefore we get to this statement
else:
    print('You have many apples!')
````

conditionals can be nested -- each "block" is defined by the indentation:

````python
'''In this program, we input a number
check if the number is positive or
negative or zero and display
an appropriate message
This time we use nested if statement'''

num = -2
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("Negative number")
````

# Iteration<a name="Iteration"></a>

`for` and `while` loops are recurrent sequence operations. That means, the computer will execute the statements inside the block (in this case `print(n)`) for a given number of times (in this case, 10 times), or until a stop criterion is fulfilled.

````python
# print the N first integer numbers:

# as a 'for loop' (given number of iterations)
N = 10
for n in range(1,N+1):
  print(n)
````

````python
# as a while loop (stop criterion)
n=1
while n <= N:
  print(n)
  n+=1
````

Often loops are needed to iterate through given data:

````python
list_1=['koala', 'cat', 'fox', 'panda', 'chipmunk', 'sloth', 'penguin', 'dolphin']

# iterate through the list and print every single item:
for x in list_1:
  print(x)
````

Here is a way to combine two lists with a for loop:

````python
list_2=['wolf', 'lion', 'ant', 'eel', 'owl', 'rat']

# iterate through the 2nd list and append all items to the first one
for x in list_2:
  list_1.append(x)

print(list_1)
````

It can be very helpful to enumerate the items from a list while iterating through your list:

````python
for i,x in enumerate(list_1):
  print(i,x)
````

Note, that enumerations are 0-based! You don't start with the first, but with the 0th element so that the number for the last element from a list of length N is N-1!

````python
print('length of list_1:', len(list_1))
````

You can also empty a list successively. Here is how to do that using a while loop.

````python
fruitsList = ['mango','apple','orange','banana','melon']

while fruitsList:
    print(fruitsList.pop(), 'is removed from the list')

print(fruitsList)
````

Iterate through dictionaries:

````python
dict_1={'mango':900, 'apple': 1100, 'orange': 2300, 'banana': 1050, 'melon': 3200}

# iterate through keys only
for key in dict_1.keys():
  print(key)

# iterate through values only
for value in dict_1.values():
  print(value)

for key, value in dict_1.items():
  print(key, value)
````

# Functions<a name="Functions"></a>


A function is a block of code that performs a specific task and which only runs when it is called.

Functions help break our program into smaller and modular chunks. As a program grows larger and larger, functions make it more organized and manageable.

Furthermore, it avoids repetition and makes the code reusable.

You can pass data, known as parameters or arguments, into a function and it returns data as a result.

````python
# only defining - no execution:
# this function takes no arguments

def hello_world():
  print("Hallo Welt! Moi, maailma!")
````

````python
# prints the the function object, not calling the function
hello_world
````

````python
# Call the function by writing () after the function name
hello_world()
````

````python
# Example function that takes one argument - a number and prints a 0 when it's even and a 1 when it's odd.

def myfunc(x):
    if x % 2 == 0:
        print(f"{x} is even")
        return 0
    else:
        print(f"{x} is odd")
        return 1
````

````python
# Call the function by writing ( argument ) after the function name
myfunc(567)
````

````python
# Example function to reverse a string

def string_reverse(str1):
    rstr1 = ''
    index = len(str1)
    while index > 0:
        rstr1 += str1[ index - 1 ]
        index = index - 1
    return rstr1

print(string_reverse('1234abcd'))
````

In the definition of your function, you might have "default values" for some of the parameters.  You can specify them in the first line.  Here, we set the default value of `b` as `None`.  If the user doesn't give a value for `b`, it gets automatically set to `None`.

````python
# default arguments

def arguments(a, b='None'):
  print('a =', a)
  print('b =', b)
  if b != 'None':
    return(a-b)
  else:
    return(b)

print(arguments(a=1))
````

# Recursion<a name="Recursion"></a>

What is recursion? It's when a function **calls** itself. It must be handled with care!

````python
# define a function that calls itself in an endless loop
def recursive_function():
  print("calling 'recursive_function'")
  recursive_function()

# this function will call itself endlessly until the system notices
# something is wrong and cancels the endless loop for you, throwing an error.
recursive_function()
````

the system keeps track of how many times the function calls itself.  It was greater than the "recursion limit", so the system throws an error.  Phew!!

We can check the current recursion limit by `getrecursionlimit()`

````python
from sys import getrecursionlimit
getrecursionlimit()
````

Used with care this kind of function can be useful, though.
e.g. to find the sum

$$ 1 + 2 + 3 + ... + n$$

We can define the $n$th triangle number $T_n = 1 + 2 + 3 + ... + n$

And notice that $T_1=1$ and $T_n = T_{n-1} + n$, which we can write as code:

````python
def triangle_number(n):
  if n == 1:
    return 1
  else:
    return triangle_number(n-1) + n
````

````python
triangle_number(10)
````

As a 2nd fun example suppose we want to draw a triangle out of smiles. This can be achieved by using recursion, too:

````python
def countdown_smilies(num):
    if num <= 0:
        print(0)
        return
    else:
        emojismiles = []
        for i in range(0, num):
            emojismiles += '😃'
        print(num, ' '.join(emojismiles))
        countdown_smilies(num - 1)
````

````python
countdown_smilies(6)
````

````python
## solution
import numpy as np
# (i)

def vowel_consonant_counter(word = 'yourstring'):
  set_of_vowels = ['a','e','i','o','u']
  v_count = 0
  c_count = 0
  for x in word:
    if x in set_of_vowels:
      v_count +=1
    else:
      c_count +=1

  return v_count,c_count


v_count,c_count = vowel_consonant_counter(word = 'aminoacid')

print('vowels in word: ', v_count)
print('consonants in word: ', c_count)


# (ii)

def vol_sphereORcylinder(radius,height = None):
  if height == None:
    print('sphere')
    V = 4./3. * np.pi * radius**3
  else:
    print('cylinder')
    V = np.pi * radius**2 * height
  return V


print(vol_sphereORcylinder(2))
print(vol_sphereORcylinder(2,1))

# (iii)

measures = [0.1,0.5,0.3,0.6,0.2,0.1]

def mean_var(measures):
  mean = 0
  for x in measures:
    mean+=x
  mean /= len(measures)
  var = 0
  for x in measures:
    var += (x-mean)**2
  var /= len(measures)

  return mean,var

mean,var =  mean_var(measures)

print(mean, np.mean(measures))
print(var, np.var(measures))

````

## Exercise 4: recursion

(i) [10 points] The famous Fibonacci numbers are defined recursively as: $F_0 = 0$, $F_1=1$ and $F_n = F_{n-1} + F_{n-2}$.
Write a recursive function that returns the $n$th Fibonacci number.

````python
## solution
def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))
````

````python
recur_fibo(4)
````

(ii) [10 points] Write a recursive function that returns the factorial of an integer number. $n! = n\cdot (n-1)\cdot (n-2)\cdot ...\cdot 1$

````python
## solution
def factorial(num):
    if (num == 0):
        return 1
    else:
        return num * factorial(num - 1)
````

TK: Cool example featuring lru_cache for speeding up the fibo function:

````python
from functools import lru_cache
import time
import numpy as np
import matplotlib.pyplot as plt

def fib_without_cache(n):
    if n < 2:
        return n
    return fib_without_cache(n-1) + fib_without_cache(n-2)

@lru_cache()
def fib_with_cache(n):
    if n < 2:
        return n
    return fib_with_cache(n-1) + fib_with_cache(n-2)

fast, slow = [], []
X = np.arange(10, 35, 5)
for N in X:

	begin = time.time()
	fib_without_cache(N)
	end = time.time()
	slow.append(end-begin)

	begin = time.time()
	fib_with_cache(N)
	end = time.time()
	fast.append(end-begin)


plt.plot(X, fast, label = 'with decorator')
plt.plot(X, slow, label = 'without decorator')
plt.legend()
plt.yscale('log')
plt.ylabel('computation time (s)')
plt.xlabel('Fibonacci N')
plt.show()
````
