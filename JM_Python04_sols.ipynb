## linear regression

Linear regression is a method to test whether an observed measure depends linearly on another quantity.

Let $x$ be the independent quantity (sometimes called "predictor" or "input") and let the depending quantity be $y$. Then, our assumption or model on how $y$ depends on $x$ is given by:

$ y(x) = ax + b $.

In the following section we will see how you can fit your model to data and how to estimate the quality of your assumptions.

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def linear_model(X, a, b):
    return [a*z + b for z in X]
```

```python
# Function to calculate the exponential with constants a and b
def exponential_model(x, a, b):
    return a*np.exp(b*x)
```

### B) Waiting times (40 pts)

From the lecture this morning, we learned that waiting times are exponentially distributed.  Below is the code from the lecture to reproduce the waiting times together with the corresponding histogram.

NOTE:  You might get a nicer picture with more bins, but there might be bins with no values.  This will cause problems when dealing with exponentials or logs!  For the purpose of the regression, it's better to use fewer bins to avoid y-values that are 0.

For the same reason it might be helpful to cut your data at larger $x$ values when $y$ tends to get small. Whenever there is a bin with zero values is produced it might run into an error.  

```python
# ----------------------------------------------
# Import required modules
# ----------------------------------------------
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------
# Define a function that returns 0 (1) with
# probabilities p (1-p)
# ----------------------------------------------

def get_random_number_01(p):
  """
  Returns 0 with probability p and 1 with probability 1-p

  Parameters:
  -----------
  p : float [0,1]

  Returns:
  -----------
  0 or 1: integers
  """

  # draw random number [0,1)
  uniform_random_variable = rnd.random()

  # convert into 0/1 with probability p
  if uniform_random_variable < p:
    return 0
  return 1

# ----------------------------------------------
# Define a function that returns the waiting
# until an event with given probability occurs
# ----------------------------------------------

def get_waiting_time(probability_of_1 = 0.1):
  """
  Function that returns the number of events that I have to wait
  until the first '1' arrives, when at every event '1' occurs with probability p

  Parameters:
  -----------
  probability_of_1: float [0,1]

  Returns:
  --------
  number_of_steps_waited: integer

  """

  # initialize counter to 1
  number_of_steps_waited = 1

  # increase counter by one as long as only '0' are drawn:
  while(get_random_number_01(1-probability_of_1) == 0):
    number_of_steps_waited += 1

  return number_of_steps_waited


# ----------------------------------------------
# Perform the simulation
# ----------------------------------------------

# Simulate a list of waiting times with a given probability:
number_of_events = 100000
probability_of_1 = 0.05

waiting_time_list = [get_waiting_time(probability_of_1) for i in range(number_of_events)]

# compare mean and variance of the simulation with theory:
print("Simulation mean     = %s [Theory: %s)" % (np.mean(waiting_time_list), 1./probability_of_1))
print("Simulation variance = %s [Theory: %s]" % (np.var(waiting_time_list), 1./probability_of_1**2))


# ----------------------------------------------
# Plotting a histogram of the simulation results
# and the theoretical exponential curve
# ----------------------------------------------

# creating x and y values for the theoretical curve
x_theory = np.arange(0,max(waiting_time_list),1)
y_theory = [probability_of_1*np.exp(-probability_of_1*x) for x in x_theory]

plt.clf() # clear previous plot, just in case

# Plot histograms of the dataset
h = plt.hist(waiting_time_list,
             histtype='bar',
             bins = 50, density = True,
             color = 'r', lw = 2, rwidth = 0.9,
             label = 'Data', align='mid')
```

```python
# solution
X = h[1][1:]
Y = h[0]

X_cutoff = []
Y_cutoff = []

for i,y in enumerate(Y):
  if y > 1e-8:
    X_cutoff.append(X[i])
    Y_cutoff.append(y)
  else:
    break
```

```python
# Function to calculate the exponential with constants a and b
# Note that they should introduce a minus in the exp
def exponential_model(x, a, b):
    return a*np.exp(-b*x)
```

```python
# give your model and your data to the method
popt, pcov = curve_fit(exponential_model, X_cutoff, Y_cutoff)
perr = np.sqrt(np.diag(pcov))

# and get back the estimated parameter values
# print(popt)
print('parameters from exponential model: a=(%.3f +\- %.3f), b=(%.3f  +\- %.3f)' %(popt[0],perr[0],popt[1],perr[1]))

# plug into your defined model function the optimized parameters
optimized_fit = exponential_model(X,popt[0],popt[1])

# plotting the data
plt.scatter(X,Y,c='gray', marker='o', edgecolors='k', s=18)
# plotting the optimized function
plt.plot(X,optimized_fit,'k',label = 'parameters: a=%.1f, b=%.1f' %(popt[0],popt[1]))
plt.legend(fontsize = 14)
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
```

```python
# X stays the same but we take the log for Y

logY = [np.log(y) for y in Y_cutoff]

plt.scatter(X_cutoff,logY,c='gray', marker='o', edgecolors='k', s=18)
plt.xlabel('x',fontsize=14)
plt.ylabel('log y',fontsize=14)
```
