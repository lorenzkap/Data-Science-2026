```python
import string
import random as rnd

# generate 2 random sequences of random letters of length N=10 as lists
#
N = 10
all_letters = string.ascii_lowercase

rnd_string_1 = rnd.sample(all_letters,N)
rnd_string_2 = rnd.sample(all_letters,N)

print('all the letters: ', all_letters)

print('first random sequence: ', rnd_string_1)
print('second random sequence: ', rnd_string_2)
```

```python
## solution

# (i)

intersection = set(rnd_string_1) & set(rnd_string_2)
print('intersection: ', intersection)

union = set(rnd_string_1) | set(rnd_string_2)
print('union: ', union)

diff_1 = set(rnd_string_1) - set(rnd_string_2)
print('set 1 minus set 2: ', diff_1)

diff_2 = set(rnd_string_2) - set(rnd_string_1)
print('set 1 minus set 2: ', diff_2)

# (ii)

expected_union = diff_1 | diff_2 | intersection
print('Both sets are equal?', expected_union == union)

# (iii)

# if set 2 is a complete subset from set 1 than we have
# union(set_1,set_2) = set_1
# or
# set_1 - set_2 = intersection(set_1,set_2)
# and there are more possibilities


```
